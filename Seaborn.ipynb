```python
import seaborn as sns
df1= sns.load_dataset("penguins") ## calling dataset directly from github
df1
```

```python
## Task is to create a line plot
## on x axis we have bill length and y axis have flipper length
```

```python
import matplotlib.pyplot as plt
sns.lineplot(x = "bill_length_mm",y="flipper_length_mm", data = df1)
plt.show()
```

## catagoriacal data clissification

```python
sns.lineplot(x = "bill_length_mm",y="flipper_length_mm", data = df1, hue = "sex", style = "sex", markers = ["o",">"])
plt.show()
```

# Bar plot in Seaborn Library

```python
import pandas as pd
```

```python
# lets take only island name from the data set above
df1["island"]
```

```python
sns.barplot(x = df1.island, y = df1.bill_length_mm)
```

```python
sns.barplot(x = "island",y="bill_length_mm",data = df1, hue = "sex")
plt.show()
```

## Ordering the bar

```python
order1 = ["Dream",'Biscoe','Torgersen']
sns.barplot(x = "island",y = "bill_length_mm", data = df1, hue = "sex", order = order1)
plt.show()
```

## Palette color change and Saturation

```python
order1 = ["Dream",'Biscoe','Torgersen']
sns.barplot(x = "island",y = "bill_length_mm", data = df1, hue = "sex", order = order1, 
           palette = "gist_heat_r", saturation = 0.5)
plt.show()
```

# Histogram in Seaborn Library

```python
df1
```

```python
sns.displot(df1["flipper_length_mm"])
plt.show()
```

## creating proper interval on x axis using bins parameter

```python
bin = [170,180,190,200,210,220,230]
sns.displot(df1["flipper_length_mm"], bins = bin)
plt.show()
```

## adding colors

```python
bin = [170,180,190,200,210,220,230]
sns.displot(df1["flipper_length_mm"], bins = bin, color = 'g')
plt.show()
```

## Kernal density (kde)

```python
bin = [170,180,190,200,210,220,230]
sns.displot(df1["flipper_length_mm"], bins = bin, kde=True)
plt.show()
```
