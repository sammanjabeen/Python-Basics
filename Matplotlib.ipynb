# Matplotlib Library

```python
import matplotlib.pyplot as plt
```

# Linear Graph

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.plot(x,y)
```

```python
plt.show()
```

## Change color of the graph

```python
x = [5,2,9,4,7,10,15,5,10]
y = [5,6,7,8,4,2,6,8,9]
c = 'r'
plt.plot(x,y,c)
plt.show()
```

## using dots 

```python
x = [5,2,9,4,7,10,15,5,10]
y = [5,6,7,8,4,2,6,8,9]
c = 'r'
plt.plot(x,y,c, marker = 'o')
plt.show()
```

## seperate graphs of x and y

```python
x = [5,2,9,4,7,10,15,5,10]
y = [5,6,7,8,4,2,6,8,9]
c = 'r'
plt.plot(x ,marker = 'o')
plt.plot(y, marker = '>')
plt.show()
```

## Bar Graph

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.bar(x,y)
plt.show()
```

### changing colors of bars 

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = ['r','g','b','y']
plt.bar(x,y, color = c)
plt.show()
```

### Graphs using arrasy

```python
import numpy as np
```

```python
x = np.array(['A','B','C','D'])
y = np.array([3,8,5,10])

plt.bar(x,y, color = 'y') # h for horizontal
# we can add colors by adding color codes
plt.show()
```

## changing width of the graphs

```python
x = np.array(['A','B','C','D'])
y = np.array([3,8,5,10])

plt.bar(x,y, color = 'y', width = 0.5)
plt.show()
```

## balancing the height of bar graph

```python
x = np.array(['A','B','C','D'])
y = np.array([3,8,5,10])

plt.barh(x,y, color = 'r', height = 0.5)
plt.show()
```

# Scatter plot 

```python
x = [1,2,3,4,5,6]
y = [2,4,12,8,10,6]

plt.scatter(x,y)
plt.show()
```

## adding title of the graph

```python
x = [1,2,3,4,5,6]
y = [2,4,12,8,10,6]
plt.title("Scatter Plot")
plt.scatter(x,y)
plt.show()
```

 ## adding lables on the x axis and y axis of the scatter plot 

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,4,12,8,10,6]
plt.title("Scatter Plot")
plt.xlabel("value 1")
plt.ylabel("value 2")
plt.scatter(x,y, color = 'r')
plt.show()
```

## changing colors of every point

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,4,12,8,10,6]
plt.title("Scatter Plot")
plt.xlabel("value 1")
plt.ylabel("value 2")
c = ['red','green','blue','yellow','white','black']
plt.scatter(x,y, color = c)
plt.show()
```

## changing the sizeof the scatter plot

```python
x = [1,2,3,4,5,6]
y = [2,4,12,8,10,6]
plt.title("Scatter Plot")
plt.xlabel("value 1")
plt.ylabel("value 2")
c = ['red','green','blue','yellow','white','black']
plt.scatter(x,y, color = c , s = 150)
plt.show()
```

# Pie Chart of the MatplotLib

```python
x = [10,20,30,40]
y = ['English', 'Urdu', 'Maths', 'Science']

# creating piechart

plt.pie(x)
plt.show()
```

## adding labels

```python
x = [10,20,30,40]
y = ['English', 'Urdu', 'Maths', 'Science']

# creating piechart

plt.pie(x, labels=y)
plt.show()
```

## changing the colors of the pie chart

```python
x = [10,20,30,40]
y = ['English', 'Urdu', 'Maths', 'Science']
c = ['yellow','magenta', 'brown', 'aqua']
# creating piechart

plt.pie(x, labels=y, colors = c)
plt.show()
```

## adding ledgend on the pie chart

```python
x = [10,20,30,40]
y = ['English', 'Urdu', 'Maths', 'Science']
c = ['yellow','magenta', 'brown', 'aqua']
# creating piechart
fname1 = r'plot.jpg' # creating name for the image
plt.pie(x, labels=y, colors = c)
plt.legend()
plt.savefig(fname1) # saving image 
plt.show()
```
